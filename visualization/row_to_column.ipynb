```python
# create empty lists to store values
row_values_list = []
binary_values_list = []

# loop through the rows of the dataframe
for index, row in result.iterrows():
    # get the values for columns 2 onwards (index 1 onwards)
    row_values = row.values[1:-1]
    # add the row values to the list
    row_values_list.append(row_values)
    # repeat the binary value for each column in the row
    binary_values = [row['understanding']] * len(row_values)
    binary_values_list.append(binary_values)

# create a new dataframe with each element in the lists in row_values_list as a separate row
# and add a new column for the binary values
new_df = pd.DataFrame({
    'Log_RT': [val for sublist in row_values_list for val in sublist],
    'Understanding': [val for sublist in binary_values_list for val in sublist]
})

print(new_df)
```
